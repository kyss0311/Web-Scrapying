```python
res= requests.get("http://httpbin.org/get", timeout = 5)
```

```python
# 載入網頁時間超過五秒就重新載入
for i in range(3):
    try:
        res= requests.get("http://httpbin.org/get", timeout = 5)
        break
    except:
        print("timeout")
```

```python
# Requests 進階用法 -Cookies
res= requests.get("https://unsplash.com/")
print(res.cookies.get_dict())
```

```python
my_cookies = { "add_new": "some_data" }
res= requests.get("http://httpbin.org/get", cookies = my_cookies )
```

```python
print(res.text)
```

```python
#  Session 會話  Session能保留每次請求後的參數，讓請求之間具有連貫性
s = requests.Session()
```

```python
s.get('https://httpbin.org/cookies/set/sessioncookie/123456789')
r = s.get('https://httpbin.org/cookies')
```

```python
print(r.text)
```

```python
# 忽略SSL憑證檢查 (對於HTTPS網站，憑證不合格會導致程式中斷)
res= requests.get("https://httpbin.org/get", verify = False)
```

```python
# 傳遞參數
payload = {'key1': 'value1', 'key2': 'value2'}
res= requests.get("http://httpbin.org/get", params = payload)
```

```python
print(res.text)
```

```python
#  自訂Requests Headers
my_headers = {'user-agent': 'my-app/0.0.1'}
res= requests.get("https://httpbin.org/get", headers = my_headers)
```
