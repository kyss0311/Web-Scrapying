```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from pyquery import PyQuery as pq
import chromedriver_autoinstaller

from selenium.webdriver.support.ui import WebDriverWait
from time import sleep

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

from random import randint, choice

chromedriver_autoinstaller.install(cwd=True)  #cwd = True 下載至當前資料夾
```

## 無頭模式(Headless)
讓chrome在背景執行 <br>
效能會更好

```python
opts = webdriver.ChromeOptions()
opts.add_argument("--headless")
driver = webdriver.Chrome(options=opts)
```

```python
driver.get("https://www.google.com.tw/imghp")
```

#### 輸入 然後按 ENTER

```python
keywor_input = driver.find_element(By.CSS_SELECTOR,'form[role="search"] textarea')
keywor_input.clear()
keywor_input.send_keys("cats")
keywor_input.send_keys(Keys.ENTER)
# 等待直到出現搜尋結果的圖片
WebDriverWait(driver, 5).until(lambda driver: driver.find_element(By.CSS_SELECTOR, '#rso div[class="wIjY0d jFk0f"] div>h3>a')) 
```

#### 捲動

```python
y_max = 0
while True:
    body = driver.find_element(By.CSS_SELECTOR, 'body>div:nth-of-type(3)')
    yOffset = body.size['height']
    # 如果已經不能在往下就跳脫迴圈
    if y_max == yOffset:
        break
    y_max = yOffset
    
    driver.execute_script("window.scrollTo(0, {})".format(yOffset))  # 捲到頁面高度的最低
    print(yOffset)
    # driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")  # 捲到底
    sleep(3)
```

### 找出圖片

```python
items = driver.find_elements(By.CSS_SELECTOR, '#rso div[class="wIjY0d jFk0f"] div>h3>a') 
```

```python
items
```

### 從最後十張照片隨機選一張 並點擊圖片

```python
items[-randint(1, 10)].click()

# 等待直到出現搜尋相關的圖片
WebDriverWait(driver, 5).until(
    lambda driver: driver.find_element(By.CSS_SELECTOR, '#Sva75c c-wiz c-wiz div[role="list"] [role="listitem"]>h3+a img')) 
```

### 找出該圖片底下的相關內容

```python
imgs = driver.find_elements(By.CSS_SELECTOR, '#Sva75c c-wiz c-wiz div[role="list"] [role="listitem"]>h3+a img')
```

```python
len(imgs)
```

#### 找出相關內容的資訊

```python
for img in imgs:
    print(img.get_attribute("alt"))
    print(img.get_attribute("src"))
```
